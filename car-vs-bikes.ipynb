```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader
import torchvision
from torchvision import transforms
from torchvision.datasets import ImageFolder
```

```python
transforme = transforms.Compose([transforms.ToTensor(), transforms.Resize((100,100))])
```

```python
#process the images with labels using the class folder
train_data = ImageFolder('D:\Dataset\Training Data\cvb\Car-Bike-Dataset', transform= transforme)
```

```python
train_data
```

```python
import numpy as np
```

```python
np.unique(train_data.targets)
```

```python
input_layer = 100 * 100 * 3
```

```python
train_loader = DataLoader(train_data, batch_size = 60,shuffle=True)
```

```python
flatten = nn.Flatten()
model = nn.Sequential(nn.Linear(input_layer, 500),
                      nn.ReLU(),
                      nn.Linear(500,500),
                      nn.ReLU(),
                      nn.Linear(500,2))
```

```python
loss_fn = nn.CrossEntropyLoss()
optim = torch.optim.Adam(model.parameters(),lr = 0.00001)
```

```python
import matplotlib.pyplot as plt
```

```python
for idx, idy in train_loader:
    idx = flatten(idx)
    print(idx.shape)
    outputs = model(idx)
    loss = loss_fn(outputs, idy)
    print(loss)
    optim.zero_grad()
    loss.backward()
    optim.step()
```
